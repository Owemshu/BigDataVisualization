```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei'] #设置字体，设置了字体后，负号会变成乱码
plt.rcParams['axes.unicode_minus'] = False #让负号的乱码正常显示

data = pd.read_excel('data/USD.xlsx').fillna(0)
data
```

```python
LiquiditySupport = data.iloc[0, 1:].to_numpy()
CreditCreation = data.iloc[1, 1:].to_numpy()
DirectLongTermLending = data.iloc[2, 1:].to_numpy()
EquitySupport = data.iloc[3, 1:].to_numpy()
HealthAndIncomeSupport = data.iloc[4, 1:].to_numpy()
InternationalAssistanceReceived = data.iloc[5, 1:].to_numpy()
AllPolicyInput = LiquiditySupport + CreditCreation + DirectLongTermLending + EquitySupport + HealthAndIncomeSupport + InternationalAssistanceReceived
```

```python
label_title = ["2020-03-20", "2020-04-04", "2020-04-18", "2020-05-01", "2020-05-15", "2020-05-29", "2020-06-13", "2020-07-27", "2020-08-10", "2020-08-24", "2020-09-07", "2020-09-21", "2020-10-05", "2020-10-19", "2020-11-02", "2020-11-16", "2020-11-30", "2020-12-18", "2021-01-11", "2021-01-25", "2021-02-08", "2021-02-22", "2021-03-08", "2021-03-22", "2021-04-26", "2021-05-31", "2021-06-28", "2021-07-26", "2021-08-30", "2021-09-27", "2021-10-25", "2021-11-15"]
# 流动性支持、信贷创造、直接长期贷款、股本支持、医疗和收入支持以及提供国际援助
plt.bar(label_title, LiquiditySupport, label='流动性支持')
plt.bar(label_title, CreditCreation, bottom=LiquiditySupport, label='信贷创造')
plt.bar(label_title, DirectLongTermLending, bottom=LiquiditySupport+CreditCreation, label='直接长期贷款')
plt.bar(label_title, EquitySupport, bottom=LiquiditySupport+CreditCreation+DirectLongTermLending, label='股本支持')
plt.bar(label_title, HealthAndIncomeSupport, bottom=LiquiditySupport+CreditCreation+DirectLongTermLending+EquitySupport, label='医疗和收入支持')
c = plt.bar(label_title, InternationalAssistanceReceived, bottom=AllPolicyInput-InternationalAssistanceReceived, label='提供国际援助')

plt.legend(loc='lower right')
plt.title('图5: 抗击新冠肺炎疫情一揽子计划投入金额堆积条形图 (2020 年 ~ 2021 年)')
plt.xticks(rotation=60)
plt.ylabel('投入金额(10亿美元)')
```
