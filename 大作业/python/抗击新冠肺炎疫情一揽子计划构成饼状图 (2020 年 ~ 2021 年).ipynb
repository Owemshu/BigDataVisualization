```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei'] #设置字体，设置了字体后，负号会变成乱码
plt.rcParams['axes.unicode_minus'] = False #让负号的乱码正常显示

data = pd.read_excel('data/USD.xlsx').fillna(0)
data
```

```python
LiquiditySupport = data.iloc[0, 1:].to_numpy()[-1]
CreditCreation = data.iloc[1, 1:].to_numpy()[-1]
DirectLongTermLending = data.iloc[2, 1:].to_numpy()[-1]
EquitySupport = data.iloc[3, 1:].to_numpy()[-1]
HealthAndIncomeSupport = data.iloc[4, 1:].to_numpy()[-1]
InternationalAssistanceReceived = data.iloc[5, 1:].to_numpy()[-1]
AllPolicyInput = LiquiditySupport + CreditCreation + DirectLongTermLending + EquitySupport + HealthAndIncomeSupport + InternationalAssistanceReceived

distribution = [LiquiditySupport, CreditCreation, DirectLongTermLending, EquitySupport, HealthAndIncomeSupport, InternationalAssistanceReceived] / AllPolicyInput
distribution
```

```python
plt.pie(x=distribution, explode=[0,0,0,0,0,1],
        labels=['流动性支持', '信贷创造', '直接长期贷款', '股本支持', '医疗和收入支持', '国际援助'],
        autopct='%.2f%%', pctdistance=0.8, labeldistance=1.1,
        counterclock=False, wedgeprops={'linewidth': 1.5}, textprops={'fontsize':8, 'color':'black'})
plt.title('图 6: 抗击新冠肺炎疫情一揽子计划构成饼状图 (2020 年 ~ 2021 年)')
```
