```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei'] #设置字体，设置了字体后，负号会变成乱码
plt.rcParams['axes.unicode_minus'] = False #让负号的乱码正常显示

gdp = pd.read_excel('data/GDP.xls').iloc[5:, 1:]
gdp
```

```python
label_title = ["2019Q1", "2019Q2", "2019Q3", "2019Q4", "2020Q1", "2020Q2", "2020Q3", "2020Q4", "2021Q1", "2021Q2", "2021Q3", "2021Q4", "2022Q1", "2022Q2", "2022Q3"]

plt.plot(label_title, gdp['国内生产总值季度累计同比增长（%）'], linewidth=2, label='国民生产总值')
plt.plot(label_title, gdp['第一产业增加值季度累计同比增长（%）'], 'g--', label='第一产业', linewidth=0.5)
plt.plot(label_title, gdp['第二产业增加值季度累计同比增长（%）'], 'b--', label='第二产业', linewidth=0.5)
plt.plot(label_title, gdp['第三产业增加值季度累计同比增长（%）'], 'y--', label='第三产业', linewidth=0.5)
plt.legend(loc='lower right')
plt.ylim(-10, 25)
plt.title('图1: 国内生产总值及个产业增长率季度变化曲线 (2019 年 ~ 2022 年)')
plt.xticks(rotation=60)
plt.ylabel('同比(%)')
plt.axhline(0, color='r')
```
