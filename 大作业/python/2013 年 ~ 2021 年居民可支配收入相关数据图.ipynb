```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei'] #设置字体，设置了字体后，负号会变成乱码
plt.rcParams['axes.unicode_minus'] = False #让负号的乱码正常显示

data = pd.read_csv('data/国内经济人均收入年度数据.csv').iloc[:, 1:].to_numpy()
data
```

```python
date_label = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']

def roll(x):
    l = []
    for i in np.arange(len(x)-1, -1, -1):
        l.append(x[i])
    return l

pcdi = roll(data[0, :] / 10000)
pcdi_group_rate = roll(data[1, :])
pcdi_median = roll(data[2, :] / 10000)
```

```python
plt.bar(date_label, pcdi, label='居民人均可支配收入')
plt.plot(date_label, pcdi_group_rate, 'r', label='居民人均可支配收入比上年增长')
plt.bar(date_label, pcdi_median, label='居民人均可支配收入中位数')
plt.legend(loc='center left')
plt.title('图 8: 2013 年 ~ 2021 年居民可支配收入相关数据图')
```
