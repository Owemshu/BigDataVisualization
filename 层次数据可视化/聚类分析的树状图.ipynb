```python
import matplotlib.pyplot as plt
import pandas as pd
# Import normalize
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
df = pd.read_csv('data/eurovision-2016.csv').fillna(0)

scores = pd.crosstab(index=df['From country'], columns=df['To country'], values=df['Televote Points'],
                     aggfunc='first').fillna(12)
samples = scores.values
country_names = list(scores.index)

# Calculate the linkage: mergings

#mergings = linkage(samples, method='single')
mergings = linkage(samples, method='weighted')
fig = plt.figure(figsize=(8, 6))
# Plot the dendrogram
dendrogram(mergings,
           labels=country_names,
           leaf_rotation=90,
           leaf_font_size=10,
           )
plt.title('Tree diagram of cluster analysis by center of gravity method')
```

```python
df = pd.read_csv('data/eurovision-2016.csv').fillna(0)

scores = pd.crosstab(index=df['From country'], columns=df['To country'], values=df['Televote Points'],
                     aggfunc='first').fillna(12)
samples = scores.values
country_names = list(scores.index)

# Calculate the linkage: mergings

#mergings = linkage(samples, method='single')
mergings = linkage(samples, method='average')
fig2 = plt.figure(figsize=(8, 6))
# Plot the dendrogram
dendrogram(mergings,
           labels=country_names,
           leaf_rotation=90,
           leaf_font_size=10,
           )
plt.title('The class averaging method draws a tree diagram of cluster analysis')
```
